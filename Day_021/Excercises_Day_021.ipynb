```python
# Create a class called Statistics
# and create all the functions that do statistical calculations as methods for the Statistics class
import math
from collections import Counter

class Statistics:
    def __init__(self,data):
        self.data = data


    def count(self):
        return len(self.data)


    def sum(self):
        total=0
        for num in self.data:
            total+=num
        return total


    def min(self):
        sorted_data=sorted(self.data)
        min=sorted_data[0]
        return min


    def max(self):
        sorted_data=sorted(self.data)
        max=sorted_data[-1]
        return max


    def range(self):
        sorted_data=sorted(self.data)
        range = sorted_data[-1]-sorted_data[0]
        return range


    def mean(self):
        total=0
        for num in self.data:
            total+=num
        mean = total/len(self.data)
        return math.ceil(mean)


    def median(self):
        sorted_data=sorted(self.data)
        if len(sorted_data)%2==0:
            median = sorted_data[int((len(sorted_data))-1/2)]
            return median
        else:
            median = (sorted_data[int((len(sorted_data))/2)] + sorted_data[int(((len(sorted_data))/2)-1)])/2
            return math.ceil(median)


    def mode(self):
        counter = Counter(self.data)
        mode = counter.most_common(1)
        return mode


    def std(self):
        total=0
        for num in self.data:
            total+=num
        mean = total/len(self.data)
        distance_from_mean = [(x-mean)**2 for x in self.data]

        sum_of_distance = 0
        for dist in distance_from_mean:
            sum_of_distance+=dist

        std = math.sqrt(sum_of_distance/len(self.data))
        return round(std,1)


    
    def var(self):
        total=0
        for num in self.data:
            total+=num
        mean = total/len(self.data)
        distance_from_mean = [(x-mean)**2 for x in self.data]

        sum_of_distance = 0
        for dist in distance_from_mean:
            sum_of_distance+=dist

        var = sum_of_distance/len(self.data)
        return round(var,1)


    def freq_dist(self):
        counter = Counter(self.data)
        freq = counter.most_common(len(self.data))
        return freq


        

       


ages = [31, 26, 34, 37, 27, 26, 32, 32, 26, 27, 27, 24, 32, 33, 27, 25, 26, 38, 37, 31, 34, 24, 33, 29, 26]
data=Statistics(ages)
print('Count: ',data.count())
print('Sum: ',data.sum())
print('Min: ',data.min())
print('Max: ',data.max())
print('Range:',data.range())
print('Mean: ',data.mean())
print('Median: ',data.median())
print('Mode: ',data.mode())
print('Standard Deviation: ',data.std())
print('Variance: ',data.var())
print('Frequency Distribution: ',data.freq_dist())
```

```python
# Create a class called PersonAccount. It has firstname, lastname, incomes, expenses properties
# and it has total_income, total_expense, account_info, add_income, add_expense and account_balance methods.
# Incomes is a set of incomes and its description. The same goes for expenses.

class PersonAccount:
    def __init__(self,firstname,lastname,incomes,expenses):
        self.firstname=firstname
        self.lastname=lastname
        self.incomes=incomes
        self.expenses=expenses


    def total_income(self):
            incomes = self.incomes.values()
            total_income=0
            for income in incomes:
                total_income+=income
            return total_income


    def total_expense(self):
            expenses = self.expenses.values()
            total_expense=0
            for expense in expenses:
                total_expense+=expense
            return total_expense


    def account_info(self):
            return f'Account Name : {self.firstname} {self.lastname}    Incomes: {self.incomes}     Expenses: {self.expenses}'


    def add_income(self,description,income):
            self.incomes[description] = income


    def add_expense(self,description,expense):
            self.expenses[description] = expense


    def account_balance(self):
            income = self.total_income()
            expense = self.total_expense()
            balance = income-expense
            return f'The account balnce is : {balance}'


# Creating instances of PersonAccount
my_account = PersonAccount('Abubakar','Auwal',
        {'Salary':50, 'Wages':10 }, {'shopping':5,}
        )
print(my_account.account_info())
print(my_account.add_income('Areas',100))
print(my_account.add_expense('Nepa bill',50))
print(my_account.account_info())
print(my_account.total_income())
print(my_account.total_expense())
print(my_account.account_balance())





```
